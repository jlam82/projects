```python
import pandas as pd

df = pd.read_csv("yaws_csv_RAW1.csv") # loading the .csv file made from "Antoine Coefficient OCR reading 1.ipynb"
df
```

```python
df.drop(labels=["ID"], axis=1, inplace=True) # the ID column serves us no purpose
df
```

```python
import numpy as np

No = np.arange(1, 59, 1)
idx = No.searchsorted(14.1)
No = np.concatenate((No[:idx], [14.1], No[idx:]))

No_df = pd.DataFrame(No, columns=["No."])

df.update(No_df) # fixing the "No." column
df 
```

```python
df.drop(labels=["Formula"], axis=1, inplace=True) # dropping another column that serves us no purpose
df
```

```python
df.columns = ["No.", "Formula", "Name", "CAS No.", "A", "B", "C", "Tmin", "Tmax"] # the initial column naming did not match up with the ocr readings
df
```

# <u>Manual Entry Corrections:</u>

```python
df.iloc[24].Name = "dysprosium"
df.iloc[38].Name = "iodine"
df.iloc[53].Name = "neon"
df.iloc[56].Name = "oxygen"
df.iloc[57].Name = "osmium"
df
```

```python
df.drop(labels="CAS No.", axis=1, inplace=True) # also decided to drop CAS No (I think it be very rare if this info would be needed)
df
```

```python
df.iloc[18].A = "7.063" # chlorine
df.iloc[23].A = "6.315" # deuterium
df.iloc[31].A = "8.017" # gadolinium
df.iloc[34].A = "5.271" # helium-4
df.iloc[47].A = "7.706" # manganese; hard error catch, ',' instead of '.'
df
```

```python
df.iloc[38].B = "1.5570 E+3" # iodine (again)
df.iloc[44].B = "8.8846 E+3" # lithium
df.iloc[49].B = "2.8557 E+2" # nitrogen

for n in df.index:
    split = df["B"][n].split() # iterate through n instead of i
    if len(split) == 2: # from the manual correction above, all entries here must have one space
        power = f"e{split[1][2]}"
        antoine_B = "".join([split[0], power])
        df.iloc[n].B = antoine_B # all B entries are now of proper scientific notation formatting
        
df # checkpoint
```

```python
df.iloc[4].C = "271.80" # argon
df.iloc[20].C = "59.51" # chromium
df.iloc[21].C = "255.57" # cesium
df
```

```python
df.iloc[0].Tmin = "1599.85" # actinium; later iterate to remove ',' even from correct entries
df.iloc[2].Tmin = "1209.00" # aluminum
df.iloc[4].Tmin = "-189.37" # argon
df.iloc[14].Tmin = "2566.00" # graphite
df.iloc[15].Tmin = "351.85" # calcium
df.iloc[18].Tmin = "-101.03" # chlorine
df.iloc[42].Tmin = "-157.37" # krypton
df.iloc[45].Tmin = "1600.00" # lutetium
df.iloc[53].Tmin = "-248.59" # neon
df.iloc[56].Tmin = "-218.80" # oxygen
df.iloc[57].Tmin = "2600.00" # osmium

for n in df.index:
    i = df.iloc[n].Tmin
    if "," in i:
        df.iloc[n].Tmin = "".join(i.split(","))
        
df
```

```python
df.iloc[19].Tmax = "2254.85" # cobalt
df.iloc[22].Tmax = "2562.00" # copper
df.iloc[37].Tmax = "2700.00" # holmium
df.iloc[38].Tmax = "546.00" # iodine
df.iloc[41].Tmax = "1949.85" # potassium
df.iloc[52].Tmax = "3074.00" # neodynium
df.iloc[53].Tmax = "-228.75" # neon
df.iloc[55].Tmax = "1800.84" # neptunium

for n in df.index:
    i = df.iloc[n].Tmax
    if "," in i:
        df.iloc[n].Tmax = "".join(i.split(","))
        
df
```

```python
pt = pd.read_csv("periodic_table.csv") # pt for "periodic table"

for n in df.index:
    df_form = df.iloc[n].Name.capitalize()
    pt_name = list(pt["Name"])
    if df_form in pt_name:
        row = pt.loc[pt["Name"] == df_form]
        irow = row.index[0]
        sym = pt["Symbol"][irow]
        
        df.iloc[n].Formula = sym
        
df
```

```python
df.iloc[13].Formula = "aC" # https://en.wikipedia.org/wiki/Amorphous_carbon
"""now manually changing the formulas missed by the for loop above; also, mispelled 'amorphous' in the original reading..."""
df.iloc[14].Formula = "C"
df.iloc[23].Formula = "D"
df.iloc[34].Formula = "He"
df.iloc[58].Formula = "P"
df
```
