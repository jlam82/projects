```python
import pandas as pd
import numpy as np

df = pd.read_csv("yaws_csv_RAW2.csv")
df
```

```python
df.drop(labels=["0", "Column0", "Column1", "Column5"], axis=1, inplace=True)
df
```

```python
df.columns = ["No.", "Formula", "Name", "A", "B", "C", "Tmin", "Tmax"]
df
```

```python
No = list(np.arange(59, 95, 1))
No.insert(0, 58.1) # easy insertion, unlike the last "yaws_csv Cleaning1.ipynb"

No_df = pd.DataFrame(No, columns=["No."])
df.update(No_df)
df
```

```python
df.iloc[1].Name = "lead"
```

```python
df.at[1, "Name"] = "lead"
df.at[5, "Name"] = "platinum"
df.at[9, "Name"] = "rhenium"
df.at[14, "Name"] = "antimony"
df.at[22, "Name"] = "terbium"
df.at[23, "Name"] = "technetium"
df
```

```python
pt = pd.read_csv("periodic_table.csv") # pt for "periodic table"

for n in df.index:
    df_form = df.iloc[n].Name.capitalize()
    pt_name = list(pt["Name"])
    if df_form in pt_name:
        row = pt.loc[pt["Name"] == df_form]
        irow = row.index[0]
        sym = pt.at[irow, "Symbol"]
        
        df.at[n, "Formula"] = sym
        
df # this cell may need to be ran twice (for some reason)
```

```python
df.at[0, "Formula"] = "P"
df.at[16, "Formula"] = "Se"
df.at[19, "Formula"] = "Sn"
df
```

```python
df.at[14, "A"] = 5.881
df
```

```python
df.at[3, "B"] = "5.2650 E+3"
df.at[25, "B"] = "2.5785 E+4"

for n in df.index:
    split = df.at[n, "B"].split()
    if len(split) == 2: # from the manual correction above, all entries here must have one space
        power = f"e{split[1][2]}"
        antoine_B = "".join([split[0], power])
        df.at[n, "B"] = antoine_B
        
df
```

```python
df.at[15, "C"] = 45.18
df.at[28, "C"] = 200.67
df.at[35, "C"] = 318.01
df
```

```python
df.at[22, "Tmin"] = "1516.10"
df.at[32, "Tmin"] = "-190.00"
df.at[33, "Tmin"] = "1610.10"

for n in df.index:
    i = df.at[n, "Tmin"]
    if "," in i:
        df.at[n, "Tmin"] = "".join(i.split(","))
        
df
```

```python
df.at[21, "Tmax"] = "5458.00"
df.at[25, "Tmax"] = "4788.00"
df.at[31, "Tmax"] = "5555.00"

for n in df.index:
    i = df.at[n, "Tmax"]
    if "," in i:
        df.at[n, "Tmax"] = "".join(i.split(","))
        
df
```

```python
df.to_csv("yaws_csv_clean2.csv")
```
