```python
import os

folders = ["Yaws reading 1", "Yaws reading 2"]

directory = os.listdir()

paths = []
for i in folders:
    if i in directory:
        path = os.path.join(os.getcwd(), i)
        paths.append(path)
```

```python
import csv

csvs = ["yaws_csv_clean1.csv", "yaws_csv_clean2.csv"]
csv_file1, csv_file2 = open(os.path.join(paths[0], csvs[0])), open(os.path.join(paths[1], csvs[1]))

yaws_csv = open("yaws_csv.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(yaws_csv)

for i in (csv_file1, csv_file2):
    if i == csv_file2:
        heading = next(i)
        reader2 = csv.reader(i)
        for j in reader2:
            csv_writer.writerow(j)
    else:
        reader1 = csv.reader(i)
        for j in reader1:
            csv_writer.writerow(j)
    i.close() # important step here
            
yaws_csv.close()
```
